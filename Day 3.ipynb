```python
# Import key libraries and paths
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
```

```python
path = 'C:/Users/Anne Joo Marthinsen/Documents/GitHub/adventofcode'
with open('day3-engine_schematic.txt') as f:
    lines = f.readlines()
```

```python
symbols = "*@/$=-+&#%"
```

```python
def check_symbol_in_line(line):
    for symbol in symbols:
        if symbol in line:
            return True
    return False
```

```python
numbers = []
indexes = []

for i in range(len(lines)):
    print('-----')
    if i == 0:
        line = lines[i]
        next_line = lines[i+1]
        print(line)
        print(next_line)
    elif i == len(lines)-1:
        prev_line = lines[i-1]
        line = lines[i]
        print(prev_line)
        print(line)
    else:
        prev_line = lines[i-1]
        line = lines[i]
        next_line = lines[i+1]
        print(prev_line)
        print(line)
        print(next_line)

    n = len(line)
    # Find next number
    j = 0
    while j < n:
        char = line[j]
        if char.isnumeric():
            for k in range(n-j):
                next_char = line[j+k]
                if next_char.isnumeric():
                    continue
                else:
                    # Found the number
                    number = line[j:j+k]
                    index = j
                    print(f'Number is {number}')
                    #print(f'Length of number is {len(number)}')

                    #Number is at the beginning of the line
                    if j == 0:
                        start = 0
                        end = j+1+len(number)
                    elif j == n:
                        start = j-1
                        end = n
                    else:
                        start = j-1
                        end = j+1+len(number)

                    # Is the number valid?
                    # First line, no line before to check
                    if i == 0:
                        if check_symbol_in_line(line[start:end]) or check_symbol_in_line(next_line[start:end]) :
                            numbers.append(number)
                            indexes.append(index)
                    # Last line, no line after to check
                    elif i == len(lines)-1:
                        if check_symbol_in_line(prev_line[start:end]) or check_symbol_in_line(line[start:end]):
                            numbers.append(number)
                            indexes.append(index)
                    else:
                        if check_symbol_in_line(prev_line[start:end]) or check_symbol_in_line(line[start:end]) or check_symbol_in_line(next_line[start:end]):
                            numbers.append(number)
                            indexes.append(index)
                            
                    
                    j += len(number)
                    break

        j += 1
            

                

            
```

```python
tot = 0

for i in range(len(numbers)):
    tot += int(numbers[i])

print(tot)
```

```python
directions = lines[0]
print(directions)

nodes = lines[2:]
print(nodes)

current_node = 'AAA'
total_traverses = 0
```

```python
def translate_directions(directions):
    numerical_directions = []
    for letter in directions:
        if letter == 'L':
            numerical_directions.append(0)
        elif letter == 'R':
            numerical_directions.append(1)

    return numerical_directions
```

```python
numerical_directions = translate_directions(directions)
print(numerical_directions)
```

```python
nodes_dict = {}

for item in nodes:
    # Remove whitespace and newline characters
    item = item.strip()
    
    # Split the string into parts
    parts = item.split('=')
    
    # Extract key and values
    key = parts[0].strip()

    values = [value.strip() for value in parts[1][2:-1].split(',')]  # Extract values from parentheses
    
    # Add to the dictionary
    nodes_dict[key] = values

print(nodes_dict)
```

```python
def traverse_directions(current_node, directions, nodes_dict):
    num_traverses = 0
    if current_node == 'ZZZ':
        return current_node, num_traverses
    
    for direction in directions:
        current_node = nodes_dict[current_node][direction]
        num_traverses += 1
        print(current_node)

        if current_node == 'ZZZ':
            return current_node, num_traverses

    return current_node, num_traverses
```

```python
current_node = 'AAA'
total_traverses = 0

while current_node != 'ZZZ':
    current_node, num_traverses = traverse_directions(current_node, numerical_directions, nodes_dict)
    total_traverses += num_traverses
print(total_traverses)
```

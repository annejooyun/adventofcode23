```python
numbers_text = ['one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine']
numbers_int = [1, 2, 3, 4, 5, 6, 7, 8, 9]
```

```python
def find_first_number_text(line):
    for i in range(len(line)):
        front_sub_line = line[0:i]
        for j in range(9):
            if numbers_text[j] in front_sub_line:
                return numbers_int[j], i-len(numbers_text[j])
    return None, None

def find_last_number_text(line):
    for i in range(1,len(line)):
        back_sub_line = line[-i:]
        for j in range(9):
            if numbers_text[j] in back_sub_line:
                return numbers_int[j], len(line)-i
    return None, None

def find_first_number_int(line):
    for i in range(len(line)):
        if line[i].isnumeric():
            return int(line[i]), i
    return None, None

def find_last_number_int(line):
    for i in range(1, len(line) + 1):
        if line[-i].isnumeric():
            return int(line[-i]), len(line)-i
    return None, None
```

```python
tot = 0
for line in lines:    
    first_num_t, i = find_first_number_text(line)
    first_num_i, j = find_first_number_int(line)
    #print(f'first numbers and positions {first_num_t, i} and {first_num_i, j}')

    last_num_t, k = find_last_number_text(line)
    last_num_i, l = find_last_number_int(line)
    #print(f'last numbers and positions {last_num_t, k} and {last_num_i, l}')

    if text_included:
        if first_num_t == None:
            first_num = first_num_i
        elif first_num_i == None:
            fist_num = first_num_t
        elif i < j:
            first_num = first_num_t
        else:
            first_num = first_num_i

        if last_num_t == None:
            last_num = last_num_i
        elif last_num_i == None:
            last_num_i = last_num_t
        elif k > l:
            last_num = last_num_t
        else:
            last_num = last_num_i

        print(f'Final first and last numbers {first_num, last_num}')

    else:
        first_num = first_num_i
        last_num = last_num_i

    print(f'Final first and last numbers {first_num, last_num}')
    sum = int(str(first_num)+str(last_num))
    tot+=sum


print(tot)

```
